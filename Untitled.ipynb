```python
import math
def sinmoide(x):
    return 1/1+math.exp(-x)
```

```python
sinmoide(100)
```

```python
import matplotlib.pyplot as plt
import numpy as np 
def sigmoid(): 
    x=np.linspace(-10,10,100)
    y=1/(1+ np.exp(-x))
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('input')
    plt.ylabel('sigmoid output')
    plt.title('sigmoid Activation fn')
    plt.show()

sigmoid()
```
