```python
import matplotlib.pyplot as plt
import numpy as np 
def sigmoid(): 
    x=np.linspace(-10,10,100)
    y=1/(1+ np.exp(-x))
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('input')
    plt.ylabel('sigmoid output')
    plt.title('sigmoid Activation fn')
    plt.show()

sigmoid()
```

```python
def tanh_fn(): 
    x=np.linspace(-1,1,100)
    y=np.exp(x)-np.exp(-x)/np.exp(x)+np.exp(-x)
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('input')
    plt.ylabel('tanh output')
    plt.title('tanh Activation fn')
    plt.show()
    print(x)

tanh_fn()
```

```python
def Relu_fn(): 
    x=np.linspace(-1,1,100)
    def relu_fn(): 
        return np.where(x<0,0,x)
    y=np.maximum(0,x)
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('input')
    plt.ylabel('relu output')
    plt.title('relu Activation fn')
    plt.show()

Relu_fn()
```

```python
def leakyrelu_fn(): 
    x=np.linspace(-1,1,100)
    def leaky_relu(x, alpha=0.1): 
        return np.where(x>=0,x,alpha*x)
    y=leaky_relu(x)        
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('input')
    plt.ylabel('leakyrelu output')
    plt.title('leakyrelu Activation fn')
    plt.show()

leakyrelu_fn()
```

```python
def thre_fn(): 
    x=np.linspace(-1,1,100)
    def ther_relu(x, alpha=0.1): 
        return np.where(x>=0,1,0)
    y=ther_relu(x)        
    plt.plot(x,y)
    plt.grid(True)
    plt.xlabel('input')
    plt.ylabel('threshold output')
    plt.title('threshold Activation fn')
    plt.show()

thre_fn()
```

```python
import math
def softmax(x):
    e_x=np.exp(x-np.max(x))
    return e_x/np.sum(e_x,axis=0)
    
```

```python
def plot_softmax(): 
    class_label=['classA','classB','classC']
    probilities=np.array([0.2,0.3,0.5])
    plt.bar(class_label,probilities)
    plt.xlabel('class')
    plt.ylabel('probilities')
    plt.title('softmax Activation fn')
    plt.show() 
plot_softmax()
```
